## Faça como eu fiz 03

#### Chegou a hora de você testar os conhecimentos desenvolvidos durante a aula. Continuando com o projeto das laranjas/toranjas agora você deve calcular o coeficiente ângular e o linear para a reta da laranja e para a reta da toranja. Use a fórmula de mínimos quadrados para encontrar cada um.

#a = Coeficiente angular

#n -> número de elementos

#Y = pesos

#X = diametros



O coefiente angular pode ser obtido usando a equação:

$$ \hat{a} = \frac{n*Soma( X_i*Y_i ) - Soma( X_i )*Soma(Y_i)}{n Soma(X_i^2) - (Soma(X_i))^2}$$ 

```python
Y = pesos_laranja
X = diametros_laranja
n = np.size(pesos_laranja)
```

```python
X.shape
```

```python
a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2)-np.sum(X)**2)
```

```python
a
```

Coeficiente linear 

$$ \hat{b} = Media(Y_i) - \hat{a} * Media(X_i) $$

```python
b = np.mean(Y) - a*np.mean(X)
```

```python
b
```

```python
y = a*X+b
```

```python
np.linalg.norm(pesos_laranja - y)
```

```python
plt.plot(diametros_laranja, pesos_laranja)
plt.plot(diametros_laranja, y)
```

```python
Y = pesos_toranja
X = diametros_toranja
n = np.size(pesos_toranja)
```

```python
a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2)-np.sum(X)**2)
```

```python
b = np.mean(Y) - a*np.mean(X)
```

```python
y = a*X+b
```

```python
np.linalg.norm(pesos_toranja - y)
```

```python
plt.plot(diametros_toranja, pesos_toranja)
plt.plot(diametros_toranja, y)
```

#### Chegou a hora de você testar os conhecimentos desenvolvidos durante a aula. Continuando com o projeto das laranjas/toranjas, agora você deve calcular o coeficiente angular utilizando a geração de números aleatórios. Assuma que já conhece b e que este é igual a 17.

```python
#Laranjas
norma_laranjas = np.array([])
np.random.seed(40)
coeficientes_angulares = np.random.uniform(low=40, high=100, size=100)
b = 17

for i in range(100):
    norma_laranjas = np.append(norma_laranjas, np.linalg.norm(pesos_laranja - (coeficientes_angulares[i]*X+b)))

#Toranjas

norma_toranjas = np.array([])
np.random.seed(30)
coeficientes_angulares = np.random.uniform(low=40, high=100, size=100)
b = 17

for i in range(100):
    norma_toranjas = np.append(norma_toranjas, np.linalg.norm(pesos_laranja - (coeficientes_angulares[i]*X+b)))

```

```python
dados_laranjas = np.column_stack([norma_laranjas, coeficientes_angulares])
dados_toranjas = np.column_stack([norma_toranjas, coeficientes_angulares])
```

```python
np.savetxt('dados_laranjas.csv', dados_laranjas, delimiter=',')
np.savetxt('dados_toranjas.csv', dados_toranjas, delimiter=',')
```
