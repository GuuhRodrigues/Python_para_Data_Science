#### 4) Encontre o valor máximo de emissão do ano de 2021 para os dados de "Agropecuária" no Estado do Pará.

```python
colunas_info = list(emissoes_gases.loc[:,'Nível 1 - Setor':'Produto'].columns)
colunas_emissao = list(emissoes_gases.loc[:, 1970:2021].columns)
emissoes_por_ano = emissoes_gases.melt(id_vars=colunas_info, value_vars=colunas_emissao, var_name='Ano', value_name='Emissão')
```

## Desafio 3: hora da prática

#### 1) Faça um agrupamento de dados com as colunas "Estado" e "Nível 1 - Setor", obtendo a soma de emissão e armazenando o resultado em uma tabela.

```python
df_emissoes_por_ano = emissoes_por_ano.groupby(['Estado', 'Nível 1 - Setor'])['Emissão'].sum().to_frame().reset_index()
df_emissoes_por_ano
```

#### 2) Utilizando a tabela construída na atividade 1, selecione os dados referentes à "Energia" do índice "Nível 1 - Setor".

```python
dados_energia = df_emissoes_por_ano[df_emissoes_por_ano['Nível 1 - Setor'] == 'Energia']
dados_energia
```

#### 3) Utilizando a tabela construída na atividade 1, encontre a atividade econômica com valor máximo de emissão do Estado de Minas Gerais

```python
atividade_valor_emissao_MG = df_emissoes_por_ano.loc[df_emissoes_por_ano['Estado'] == 'MG'].groupby('Nível 1 - Setor')['Emissão'].max().reset_index()
```

```python
nivel_max_emissao = atividade_valor_emissao_MG.loc[atividade_valor_emissao_MG['Emissão'].idxmax(), 'Nível 1 - Setor']
print(nivel_max_emissao)
```

#### 4) Obtenha uma tabela contendo a atividade econômica com máxima emissão para cada Estado.

```python
df_atividade_maxima_emissao_estado = df_emissoes_por_ano.groupby('Estado')['Emissão'].idxmax()
df_atividade_maxima_emissao_estado = df_emissoes_por_ano.loc[df_atividade_maxima_emissao_estado].reset_index()
df_atividade_maxima_emissao_estado = df_atividade_maxima_emissao_estado.drop(columns='index')
df_atividade_maxima_emissao_estado
```

#### 5) Obtenha uma tabela contendo o Estado com máxima emissão para cada atividade econômica.

```python
df_emissoes_por_ano_outro = emissoes_por_ano.groupby(['Nível 1 - Setor', 'Estado'])['Emissão'].sum().to_frame().reset_index()
df_estado_maxima_emissao_atividade = df_emissoes_por_ano_outro.groupby('Nível 1 - Setor')['Emissão'].idxmax()
df_estado_maxima_emissao_atividade = df_emissoes_por_ano_outro.loc[df_estado_maxima_emissao_atividade].reset_index()
df_estado_maxima_emissao_atividade = df_estado_maxima_emissao_atividade.drop(columns='index')
df_estado_maxima_emissao_atividade
```

```python
df_estado_maxima_emissao_atividade.plot(kind='bar', figsize=(10, 6), ylabel='Emissão' )
```

```python
import matplotlib.pyplot as plt

# Definir os dados para o gráfico
niveis = df_estado_maxima_emissao_atividade['Nível 1 - Setor']
estados = df_estado_maxima_emissao_atividade['Estado']
emissoes = df_estado_maxima_emissao_atividade['Emissão']

# Configurar o gráfico de barras
plt.bar(niveis, emissoes)
plt.xlabel('Nível 1 - Setor')
plt.ylabel('Emissão')
plt.title('Estado com Máxima Emissão por Atividade Econômica')

# Rotacionar os rótulos do eixo x para facilitar a leitura
plt.xticks(rotation=90)

# Adicionar rótulos dos estados acima de cada barra
for i in range(len(niveis)):
    plt.text(i, emissoes[i], estados[i], ha='center')

# Exibir o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Definir os dados para o gráfico
niveis = df_estado_maxima_emissao_atividade['Nível 1 - Setor']
estados = df_estado_maxima_emissao_atividade['Estado']
emissoes = df_estado_maxima_emissao_atividade['Emissão']

# Configurar o gráfico de barras horizontais
plt.barh(niveis, emissoes)
plt.xlabel('Emissão')
plt.title('Estado com Máxima Emissão por Atividade Econômica')

# Adicionar rótulos dos estados à direita de cada barra
for i in range(len(niveis)):
    plt.text(emissoes[i], i, estados[i], va='center')

# Exibir o gráfico
plt.show()
```
